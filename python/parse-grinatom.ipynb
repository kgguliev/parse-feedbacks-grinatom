```python
items = len(driver.find_elements_by_class_name('feedback'))

good_text = []

bad_text = []

for j in range(items):
    quotes = driver.find_elements_by_class_name('feedback')
    for comments in quotes:
        comments_text = comments.find_element_by_class_name('good').text
        good_text.append(comments_text)
        
for h in range(items):
    quotes = driver.find_elements_by_class_name('feedback')
    for comments in quotes:
        comments_text = comments.find_element_by_class_name('bad').text
        bad_text.append(comments_text)
```

```python
good_text = []

bad_text = []

for pages in range(1, 11 + 1):
    driver.get('https://neorabote.net/feedback/list/company/37548/page/%s' %str(pages))
    items = len(driver.find_elements_by_class_name('feedback'))

    for j in range(items):
        quotes = driver.find_elements_by_class_name('feedback')
        for comments in quotes:
            comments_text = comments.find_element_by_class_name('good').text
            good_text.append(comments_text)
        
    for h in range(items):
        quotes = driver.find_elements_by_class_name('feedback')
        for comments in quotes:
            comments_text = comments.find_element_by_class_name('bad').text
            bad_text.append(comments_text)
```

```python
final_df = pd.DataFrame({
    'positive' : good_text,
    'negative' : bad_text
})
```
